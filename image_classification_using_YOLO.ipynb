```python
from pathlib import Path
data_path = Path("waste-classification")
data_path
```

```python
def walkthrough_dir (dir_path):
  total_image = 0;
  for dirpath ,dir_names,file_names in os.walk(dir_path):
    print(f"There are {len(dir_names)} directories  and {len(file_names)} image in {dirpath}")
    total_image +=len(file_names)

  print(f"Total image {total_image}")
```

```python
walkthrough_dir(data_path)
```
